```python
from autoscraper import AutoScraper
```

Steps: 
1. Importing Lib
2. Setting up a url and wanted list value
3. Initializing autoscraper()
4. Building a scraper which is stored in variable result1
5. 


```python
url = 'https://www.flipkart.com/search?q=on%20protein&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
wanted_list = ['Optimum Nutrition (ON) Gold Standard 100% Protein Powder - Primary Source Isolate Whey Protein','8,470 Reviews']
```

```python
scraper = AutoScraper()
result1= scraper.build(url,wanted_list)
print(result1)
```

```python
result2 = scraper.get_result_similar('https://www.flipkart.com/search?q=on%20protein&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off', grouped=True)
result2
```

```python
scraper_rule_line= scraper.set_rule_aliases({'rule_i6m5': 'Title'})
print(scraper_rule_line)
```

```python
scraper.keep_rules(['rule_i6m5'])
```

```python
scraper.save("flipkart_onProtein")
```

```python
scraper.load('flipkart_onProtein')
```
