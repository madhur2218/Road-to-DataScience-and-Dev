```python
from autoscraper import AutoScraper
```

Steps: 


```python
url = 'https://www.flipkart.com/search?q=on%20protein&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
wanted_list = ['Optimum Nutrition (ON) Gold Standard 100% Protein Powder - Primary Source Isolate Whey Protein']
```

```python
scraper = AutoScraper()
result1= scraper.build(url,wanted_list)
print(result1)
```

```python
scraper.get_result_similar('https://www.flipkart.com/search?q=on%20protein&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off', grouped=True)
```

```python
scraper.set_rule_aliases({'rule_p6fl': 'Title'})
```

```python
scraper.keep_rules(['rule_p6fl'])
```

```python
scraper.save("flipkart_onProtein")
```
