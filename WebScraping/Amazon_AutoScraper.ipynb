```python
from autoscraper import AutoScraper
```

```python
amazon_url = 'https://www.amazon.in/s?k=iphones&crid=304M35IM976NW&sprefix=iphone%2Caps%2C265&ref=nb_sb_noss_1'
```

```python
wanted_list = ["1,37,990", "Apple iPhone 15 Pro (256 GB) - Natural Titanium", '100+ bought in past month '] 
```

```python
scraper = AutoScraper()
results = scraper.build(url=amazon_url, wanted_list=wanted_list)
print(results)
```

```python
scraper.get_result_similar(amazon_url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_b5d6': 'Title'})
scraper.keep_rules(['rule_b5d6'])
scraper.save('amazon web scraping')
```
