```python
#check whether a number is negative, positive or zero.

def checkNum(num):
    if(num ==0):
        print("number is zero");
    elif (num>0):
        print("number is positive");
    else:
        print("number is negative");

num=10;
checkNum(num)
```

```python
#check whether a number is divisible by 5 and 11 or not.
def checkifDivisible(num):
    if(num%5==0 or num%11==0):
        print("divisible")
    else:
        print("not divisible");

num= int(input("enter a number"))
checkifDivisible(num)
```

```python
#check whether a year is leap year or not

def leapYear(year):
    if(year%4==0 and year%100!=0) or (year%400==0):
        print("leap year");

year = 2024;
leapYear(year)
```

```python
#check whether a character is alphabet or not.

def checkIfAlphabet(char):
    if(char.isalpha()):
        print("alphabet");
    else:
        print("not an alphabet");

char = "A";
checkIfAlphabet(char)
```

```python
#program to input any alphabet and check whether it is vowel or consonant.

def checkIfVowel(char):
    if(char.lower() or char.upper() in "aeiou"):
        print("vowel");
    else:
        print("consonant");

char = "A";
checkIfVowel(char)
```

```python
#character and check whether it is alphabet, digit or special character.

def checkIfAlphabet(char):
    if(char.isalpha()):
        print("alphabet");
    elif(char.isdigit()):   
        print("digit");
    else:
        print("special character");

char = "#";
checkIfAlphabet(char)

```

```python
# program to input week number and print week day.

def inputWeekNumber(number):
    match number:
        case 1: 
            print("monday");
        case 2:
            print("tuesday");
        case 3:
            print("wednesday");
        case 4:
            print("thursday");
        case 5:
            print("friday");
        case 6:
            print("saturday");
        case 7:
            print("sunday");
        case _:
            print("invalid week number");

weekNumber = int(input("enter week number"));
inputWeekNumber(weekNumber)

            
            
```
