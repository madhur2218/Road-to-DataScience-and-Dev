```python
## else
## The else statement executes a block of code if the condition in the if statement is False.

def calculateAge(num):
    if(num >=18):
        print("eligible to vote");
    else:
        print("not eligible to vote");

age = int(input("Enter age :"))
calculateAge(age);
```

```python
## elif
## The elif statement allows you to check multiple conditions. It stands for "else if"

def calculateAge(num):
    if(num >=18):
        print("teen");
    elif(num <18):
        print("child");
    else:
        print('adult')

age = int(input("Enter age :"))
calculateAge(age);

```

```python
## Nested Condiitonal Statements


def justifyNumber(num):
    if(num>0):
        if(num%2==0):
            print("number is positive and is even");
        else:
            print("number is positive and is odd");
    elif(num<0):
        if(num%2==0):
            print("number is negative and is even");
        else:
            print("number is negative and is odd");
    else:
        print("number is zero");

num = int(input("enter a number"));
justifyNumber(num);


```

```python
## Practical Examples

## Determine if a year is a leap year using nested condition statement

def isLeapYear(year):
    if(year%4==0):
        if(year%100==0):
            if(year%400==0):
                print("leap year");
            else:
                print("not a leap year");
        else:
            print("leap year");
    else:
        print("not a leap year");

year = int(input("Enter a year: "))
isLeapYear(year);


```

```python
## Assignment
## Simple Calculator program

num1 = float(input("Enter first number: "));
num2 = float(input("Enter second number: "));

operation = input("Enter operation");


def calculator(num1,num2,operation):
    if(operation == "+"):
        return num1+num2;
    elif(operation == "-"):
        return num1-num2;
    elif(operation == "*"):
        return num1*num2;
    elif(operation == "/"):
        if(num1==0  or num2==0):
            print('error while dividing by zero')
        elif(num1>num2):
            return num1/num2;
        else:
            return num2/num1;
    else:
        print("Invalid operation");


print(calculator(num1,num2,operation));
    
```

```python
### Determine the ticket price based on age and whether the person is a student.
# Ticket pricing based on age and student status

def checkPricing(age,isStudent):
    if(age<5):
        price = 'free';
    elif(age>=5 and age<=12):
        price = "euro 10";
    elif(age>=13 and age<=18):
        if(isStudent):
            price= "euro 15";
        else:
            price = "euro 20";
    elif(age>=19 and age<=65):
        price = "euro 25";
    else:
        price = "euro 20";
    return price;

age = int(input("Enter age: "));
isStudent = input("Are you a student? (yes/no): ").lower() == "yes";
print(checkPricing(age,isStudent));

```

```python
#Complex Example 3: Employee Bonus Calculation

def bonusCalculation(yearsOfExperience,performance):
    if(performance >=4.5):
        if(yearsOfExperience > 10):
            bonus = 20;
        elif(yearsOfExperience >=5 and yearsOfExperience <=10):
            bonus = 15;
        else:
            bonus = 10;
    elif(performance >=3.5 and performance <4.5):
        if yearsOfExperience > 10:
            bonus = 15
        elif yearsOfExperience > 5:
                bonus = 10
        else:
             bonus = 5
    else:
         bonus=0;
    return bonus;

yearsOfExperience = int(input("Enter years of service: "))
performance = float(input("Enter performance rating (1.0 to 5.0): "))
bonus = bonusCalculation(yearsOfExperience,performance);

# Calculate bonus amount
salary = float(input("Enter current salary: "))
bonus_amount = salary * bonus / 100
print("Bonus Amount: ${:.2f}".format(bonus_amount))






```
