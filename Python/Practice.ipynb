```python
class Circle:
    pi = 3.14

    # Circle gets instantiated with a radius (default is 1)
    def __init__(self, radius=1):
        self.radius = radius 
        self.area = radius * radius * Circle.pi

    # Method for resetting Radius
    def setRadius(self, new_radius):
        self.radius = new_radius
        self.area = new_radius * new_radius * self.pi

    # Method for getting Circumference
    def getCircumference(self):
        return self.radius * self.pi * 2


c = Circle()

print('Radius is: ',c.radius)
print('Area is: ',c.area)
print('Circumference is: ',c.getCircumference())
```

```python
class Circle:
    pi = 3.14  # class variable

    def __init__(self, radius):
        self.radius = radius  # instance variable

    def calArea(self):
        area = self.pi * self.radius * self.radius
        return area
    
    def calCircumfrence(self):
        circ = self.radius * self.pi * 2
        return circ

# create object
smallCircle = Circle(3)

# call method
print(smallCircle.calArea())
print(smallCircle.calCircumfrence())
```

```python
##polymorphism

class Dog:
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        return self.name + " says woof!"
    
class Cat:
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        return self.name+' says Meow!' 
    

dog = Dog("niko")
cat = Cat("flexi")

print(dog.speak())
print(cat.speak())
```

```python
##Abstraction in python

class Animal:
    
    def __init__(self,name):
        self.name=name
    
    def speak(self):
        raise NotImplementedError("subclass mush implement abstract method")
    
class Dog(Animal):

    def speak(self):
        print("woof woof")

class Cat(Animal):

    def speak(self):
        print("meow meow!!")


fido=Dog("fido")
iris=Cat("itis")
        
print(fido.speak())
print(iris.speak())
```

```python
##some special functions 

class Book:
    def __init__(self, title, author, pages):
        print("A book is created")
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return "Title: %s, author: %s, pages: %s" %(self.title, self.author, self.pages)

    def __len__(self):
        return self.pages

    def __del__(self):
        print("A book is destroyed")
```

```python
book = Book("Python Rocks!", "Jose Portilla", 159)

#Special Methods
print(book)
print(len(book))
del book
```

```python
class Line:
    
    def __init__(self,corr1,corr2):
        self.corr1=corr1
        self.corr2=corr2

    def distance(self):
        x1,y1= self.corr1
        x2,y2=self.corr2
        return ((x2-x1)**2 + (y2-y1)**2)**0.5

    def slope(self):
        x1,y1 = self.corr1
        x2,y2 = self.corr2
        return (y2-y1)/(x2-x1)
    
corr1 = (2,3)
corr2 = (8,10)

line = Line(corr1,corr2)
print(line.distance())
print(line.slope())

```
