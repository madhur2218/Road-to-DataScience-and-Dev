```python
def factorial(num):
    if(num==1 or num==0):
        return 1;
    else:
        fact=1;
        for i in range(1, num+1):
            fact*=i;
        return fact;

num1 = int(input("Enter number1: "))
print(factorial(num1))


```

```python
#Check if a variable is of specific data type

def checkDataType(value):
    if isinstance(value,int):
        print("Value is an integer")
    elif isinstance(value,float):
        print("Value is a float")
    elif isinstance(value,str):
        print("Value is a string")
    else:
        print("Value is not a number")


user_value= input("Enter a value: ");

try:
    if "." in user_value:
        checkDataType(float(user_value))
    elif user_value.isdigit():
        checkDataType(int(user_value))
    else:
        checkDataType(str(user_value))
except ValueError:
    print("Invalid input")
    
```

```python
def is_leap_year(year):
    if (year %4==0 and year %100!=0) or (year %400==0):
        return True
    else:
        return False

year = int(input("Enter a year: "))

if is_leap_year(year):
    print(f"{year} is a Leap Year ✅")
else:
    print(f"{year} is NOT a Leap Year ❌")
```

```python
#palindrome 2nd approach doubt
def is_palindrome(num):
    original = num
    reversed_num = 0
    
    while num > 0:
        digit = num % 10  # Get last digit
        reversed_num = reversed_num * 10 + digit  # Add to reversed number
        num //= 10  # Remove last digit
    
    return original == reversed_num

num = int(input("Enter a number: "))

if is_palindrome(num):
    print(f"{num} is a Palindrome ✅")
else:
    print(f"{num} is NOT a Palindrome ❌")


```
