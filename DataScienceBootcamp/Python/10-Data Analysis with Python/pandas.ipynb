```python
#Pandas
import pandas as pd
import numpy as np
```

```python
#Pandas SERIES is like 1D array like object which works just like array
data = np.array([1,2,3,4,5])  #this is one single column, this is a series from array/list implementation
series = pd.Series(data)
print(series)
```

```python
#Series from dict
data = {'a':1, "b":2}
series = pd.Series(data)
print(series)
```

```python
data = [1,2,3,4,5]
index = ['a','b','c','d','e']

series = pd.Series(data,index=index)
print(series)
```

```python
#DATAFRAME
data = [{'Name': 'madhur', 'Age': 22},{'Name': 'naman', 'Age': 22},{'Name': 'ashu', 'Age': 22}]
print(data)

df= pd.DataFrame(data)
print(df)
```

```python
data = [[1,2,3,4,5],[2,3,4,5,6]]
df= pd.DataFrame(data)
print(df)
```

```python
data = {
    'Name': ['MADHUR', 'ASHU', "NAMAN"],
    "AGE": [22,33,34]
}

df= pd.DataFrame(data)
print(df)
```
