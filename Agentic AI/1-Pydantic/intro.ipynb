```python
#Model with optional fields
from typing import Optional

class Employee(BaseModel):
    id: int
    name: str
    dapartment:str
    salary: Optional[float] = None  # Optional field with default value None    
    is_active: Optional[bool] = True  # Optional field with default value True
    
```

```python
emp1 = Employee(id=1, name="Alice", dapartment="HR")
emp2 = Employee(id=2, name="Bob", dapartment="IT", salary=75000, is_active=False) ##type casting is also possible in case of optional, required float and providede int and type caseing automatically hapened
print(emp1)
print(emp2)
```

```python
from pydantic import BaseModel
from typing import List
from typing import Optional

class Classroom(BaseModel):
    name: str
    students: List[str]
    capacity: Optional[int] = None

classroom1 = Classroom(name="Math 101", students=("Alice", "Bob", "Charlie"), capacity=30)
print(classroom1)
```

```python
##model with nested models

from pydantic import BaseModel

class Address(BaseModel):
    street: str
    city: str
    zip_code: str

class Customer(BaseModel):
    name: str
    email: str
    address: Address  ##nested model


address1 = Address(street="123 Main St", city="New York", zip_code="10001")
customer1 = Customer(name="Jane Smith", email="jane.smith@example.com", address=address1)

print(customer1)



```

```python
from pydantic import BaseModel, Field
from typing import Optional

# Field() is used to specify validation rules, default values, and aliases
class Items(BaseModel):
    name: str = Field(..., min_length=3, max_length=50)  # Name must be between 3 and 50 chars
    price: float = Field(..., gt=0)                      # Price must be > 0
    description: Optional[str] = Field(None, max_length=200)  # Optional description (max 200 chars)
    quantity: int = Field(1, ge=0)                       # Default 1, must be non-negative


# âœ… Create an instance outside the class
item1 = Items(
    name="Laptop",
    price=999.99,
    description="A high-performance laptop",
    quantity=5
)

print(item1)
```
