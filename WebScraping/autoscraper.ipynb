```python
from autoscraper import AutoScraper

url= "https://github.com/krishnaik06?tab=repositories"

wanted_list = ['The-Grand-Complete-Data-Science-Materials', '5,124']

scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar('https://github.com/krishnaik06?tab=repositories', grouped=True)
```

```python
scraper.set_rule_aliases({'rule_7f51' : 'Title' ,'rule_z5hu' : "Stars" })
scraper.keep_rules(['rule_7f51', 'rule_z5hu'])
scraper.save("github-repo-serach-krishNaik")
```

```python
scraper.load("github-repo-serach-krishNaik")
```
