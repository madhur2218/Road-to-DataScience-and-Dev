```python
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Hello, this is home <h1> hello world</h1>'

@app.route('/<name>')
def user(name):
    return f"Hello {name}!"

@app.route("/animal")
def animal():
    return "I am a lion"

@app.route("/admin")
def admin():
    # Use url_for() to get the URL of the user function with the 'name' parameter set to 'madhur'
    return redirect(url_for('user', name='madhur'))

if __name__ == '__main__':
    app.run()
```
