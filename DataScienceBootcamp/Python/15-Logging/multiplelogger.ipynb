```python
#creating multiple loggers

import logging 

##creating logger for module1
logger1 =logging.getLogger("module1")
logger1.setLevel(logging.DEBUG);

#creating logger for module2
logger2=logging.getLogger("module2") 
logger2.setLevel(logging.INFO);


for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level= logging.DEBUG,
    format = '%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt="%Y-%m-%d %H:%M:%S"
)
```

```python
##LOG MESSAGE WITH DIFFERENT LOGGERS
logger1.warning("this is a warningfrom modul1")
logger1.error("this is an error from module1")
logger2.warning("this is a warning from module2")
logger2.info("this is a INFO from modul2")
```
