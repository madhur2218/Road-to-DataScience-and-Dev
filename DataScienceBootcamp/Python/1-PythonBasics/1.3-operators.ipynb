```python
## Arithmethic Operation
a=10
b = 5

add_result=a+b  #addiiton
sub_result=a-b  #substraction 
mult_result=a*b #multiplication
div_result=a/b  #division provides the decimal value also with the result
floor_div_result=a//b ## floor division provides us non decimal value
modulus_result=a%b #modulus operation  provides us the remainder 

exponent_result=a**b ## Exponentiation


print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)
```

```python
#Comparasion operations ==, !=,>,<>
a=10;
b=20;

print(a==b) #equal to
print(a!=b) #not equal to
print(a>b) #greater than
print(a<b) #less than
print(a>=b) #greater than or equal to
print(a<=b) #less than or equal to

```

```python
#Simple calculator

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

sum_result = num1 + num2
sub_result = num1 - num2
mult_result = num1 * num2
div_result = num1 / num2

print("Sum is ", sum_result)
print("Subtraction is ", sub_result)
print("Multiplication is ", mult_result)
print("Division is ", div_result)
```
